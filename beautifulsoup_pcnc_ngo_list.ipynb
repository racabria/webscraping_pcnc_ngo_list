```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
ngo_list = []

for i in range(1, 41): # There are 40 pages to extract data from
    url = "https://www.pcnc.com.ph/certified-ngos-list/?sf_paged=" + str(i)
    html = urlopen(url)
    soup = BeautifulSoup(html, 'lxml')

    ngo_list_table = soup.find('div', class_='ngo-list').table.find_all('tr')

    for row in ngo_list_table[1:]:
        tr = []
        td_all = row.find_all('td')
        for td in td_all:
            text = BeautifulSoup(str(td), "lxml").get_text().strip().replace('View Details','')
            tr.append(text)
        ngo_list.append(tr)
    

df = pd.DataFrame(ngo_list, columns = ['Name', 'Category', 'BIR Certificate Expiry Date', 'PCNC Certificate Expiry Date', 'Website'])
df
```

```python
df.to_csv('PCNC Certified NGO list.csv') # --- export dataframe to csv file
```
